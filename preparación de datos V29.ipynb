```python
#conexion a la base de datos sqlite activity
try:
    conexion = sqlite3.connect('chembl_29.db')
    cursor= conexion.cursor()
    print('se ha establecido conexion con la base de datos', conexion)
    sql= """SELECT ac.activity_id, ac.assay_id, ac.molregno, ay.cell_id, ac.standard_relation, ac.standard_value, ac.standard_units, ac.standard_type, ay.assay_cell_type FROM activities ac INNER JOIN assays ay ON ac.assay_id=ay.assay_id WHERE ay.cell_id IS NOT NULL
   """  
    cursor.close()
except sqlite3.Error as error:
    print('Se ha producido un error: ', error)
```

```python
compound_v29 = pd.read_csv(filepath_or_buffer='compound_v29.csv', index_col=0)
comp_list = list()
for i in range(len(compound_v29)):
    smiles = compound_v29.iloc[i]['canonical_smiles']
    comp_id = compound_v29.iloc[i]['molregno']
    if '.' in smiles:
        splt_smiles = str(smiles).split('.')
        smiles = max(splt_smiles, key=len)
    comp_list.append([comp_id, smiles])
compound_v29 = pd.DataFrame(comp_list, columns=['comp_id', 'smiles'])
```

```python
compound_v29
```

```python
duplicates = compound_v29[compound_v29['smiles'].duplicated()]
compound_v29.drop_duplicates(subset='smiles', keep='first', inplace=True, ignore_index=True)
```
